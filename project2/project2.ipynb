Question: what are the top 3 dangerours (highest number of crimes) areas in Los Angeles? what are the top 3 safe (lowest number of crimes) areas in Los Angeles? 
Solution: Use grouby by by 'AREA NAME' to count the number of crimes in different areas, and store the dataframe into df4. Then, use df4.nlargest with column name "DR_NO" to select the top 3 dangerous areas in LA. Similarly, to find the top 3 safe areas in LA, we can use df4.nsmallest with column name "DR_NO".

#### 4.2 In Python, do the data science!

```python
df4=df.groupby('AREA NAME').agg('count').reset_index()
df4.nlargest(3,'DR_NO')
```

```python
df4=df.groupby('AREA NAME').agg('count').reset_index()
df4.nsmallest(3,'DR_NO')
```
